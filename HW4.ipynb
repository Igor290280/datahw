```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def owner (numb):
    for doc in documents:
        if doc['number']==numb:
            return doc['name']
            break
    return 'Документ не найден в базе'        
        
def polka (numb):
    for pol in directories.keys():
        if numb in directories[pol]:
            return pol
            break
    return 'Документ не найден в базе'

def inptnomdoc ():
    return str(input("Введите номер документа:\n"))

def inptnompol ():
    return str(input("Введите номер полки:\n"))
    
    
x=""    
while x!="q":
    
    x=str(input('Введите команду:\n'))
    if x=="p":
        y=inptnomdoc()
        docow=owner(y)
        print('Результат:')
        if docow=='Документ не найден в базе':
            print(docow)
        else:        
            print(f'Владелец документа: {docow}')
    elif x=="s":
        z=inptnomdoc()
        polow=polka(z)
        print('Результат:')
        if polow=='Документ не найден в базе':
            print(polow)
        else:        
            print(f'Документ хранится на полке: {polow}')
    elif x=="l":
        documentsnew=documents.copy()
        print('Результат:')
        for docs in documentsnew:
            docs['directory']=polka(docs["number"])
        for docs in documentsnew:
            print("№: ", docs['number'], ", тип: ", docs['type'], ", владелец: ", docs['name'], ", полка хранения: ", docs['directory'], sep="")
    elif  x=="ads":
        p=inptnompol()
        print('Результат:')
        if p in directories.keys():
            print('Такая полка уже существует. Текущий перечень полок: ', ', '.join(directories), ".", sep='')
        else:
            directories[p]=[]
            print('Полка добавлена. Текущий перечень полок: ', ', '.join(directories), ".", sep='')
    elif x=="ds":
        j=inptnompol()
        print('Результат:')
        if j in directories.keys() and directories[j]==[]:
            del directories[j]
            print('Полка удалена. Текущий перечень полок: ', ', '.join(directories), ".", sep='')
        elif j in directories.keys() and directories[j]!=[]:
            print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ', ', '.join(directories), ".", sep='')
        else:
            print('Такой полки не существует. Текущий перечень полок: ', ', '.join(directories), ".", sep='')
```
