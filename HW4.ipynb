```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def owner():
    y=inptnomdoc()
    print('Результат:')
    for doc in documents:
        if doc['number'] == y:
            a = f'Владелец документа: {doc["name"]}'
            break
        a = 'Документ не найден в базе'
    print(a)
            
        
    

        
def polka():
    z=inptnomdoc()
    print('Результат:')            
    for pol in directories.keys():
        if z in directories[pol]:
            b = f'Документ хранится на полке: {pol}'
            break
        b = 'Документ не найден в базе'
    print(b)

def polkanumb(numb):
    for pol in directories.keys():
        if numb in directories[pol]:
            return pol
            
    
    
def alldoc():
    documentsnew=documents.copy()
    print('Результат:')
    for docs in documentsnew:
        docs['directory']=polkanumb(docs['number'])
    for docs in documentsnew:
            print("№: ", docs['number'], ", тип: ", docs['type'], ", владелец: ", docs['name'], ", полка хранения: ", docs['directory'], sep="")
def funads():
    p = inptnompol()
    print('Результат:')
    if p in directories.keys():
        print('Такая полка уже существует. Текущий перечень полок: ', ', '.join(directories), ".", sep = '')
    else:
        directories[p] = []
        print('Полка добавлена. Текущий перечень полок: ', ', '.join(directories), ".", sep = '')
def funds():
    j = inptnompol()
    print('Результат:')
    if j in directories.keys() and directories[j] == []:
        del directories[j]
        print('Полка удалена. Текущий перечень полок: ', ', '.join(directories), ".", sep = '')
    elif j in directories.keys() and directories[j]!=[]:
        print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ', ', '.join(directories), ".", sep='')
    else:
        print('Такой полки не существует. Текущий перечень полок: ', ', '.join(directories), ".", sep = '')
def inptnomdoc():
    
    return str(input("Введите номер документа:\n"))

def inptnompol():
    
    return str(input("Введите номер полки:\n"))
    
    
def userinput():
    x=""    
    while x != "q":
        x = str(input('Введите команду:\n'))
        if x == "p":
            owner()
        elif x == "s":
            polka()
        elif x == "l":
            alldoc()
        elif  x == "ads":
            funads()
        elif x == "ds":
            funds()
        

userinput()        
```
