Задание 1**
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
joined = ratings.merge(movies, on='movieId', how='left')
joined.head()
```

```python
def score (rating):
    if rating <= 2:
        return 'низкий рейтинг'
    elif 2 < rating <= 4:
        return 'средний рейтинг'
    elif 4 < rating <= 5:
        return 'высокий рейтинг'
```

```python
joined['class'] = joined['rating'].apply(score)
joined.head(10)
```

Задание 2

Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

1 'Центр': ['москва', 'тула', 'ярославль'],

2 'Северо-Запад': ['петербург', 'псков', 'мурманск'],

3 'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

Результат классификации запишите в отдельный столбец region.

```python
keywords = pd.read_csv('keywords.csv')
keywords.tail(20)
```

```python
geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
            'Северо-Запад': ['петербург', 'псков', 'мурманск'],
            'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
           }



def RegionRussia (keyword):
    keyword = keyword.lower()
    region = 'undefined'
    for data in geo_data:
        for word in geo_data[data]:
            if word in keyword:
                region = data
    return region
```

```python
RegionRussia('сахалин')
```

```python
keywords['region'] = keywords['keyword'].apply(RegionRussia)
keywords[keywords['region'] != 'undefined'].head(10)
```
